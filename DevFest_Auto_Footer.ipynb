```python
import os
from PIL import Image

input_folder = "/content/drive/MyDrive/DevFest/Photos/"
output_folder = "/content/drive/MyDrive/DevFest/Output/"
footer_path = "/content/drive/MyDrive/DevFest/footer.png"

os.makedirs(output_folder, exist_ok=True)

footer = Image.open(footer_path).convert("RGBA")
footer_width, footer_height = footer.size

for filename in os.listdir(input_folder):
    if filename.lower().endswith(("jpg", "jpeg", "png")):
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path).convert("RGBA")

        # RESIZE IMAGE TO MATCH FOOTER WIDTH, MAINTAIN ASPECT RATIO
        scale = footer_width / img.width
        new_img_width = footer_width
        new_img_height = int(img.height * scale)
        img_resized = img.resize((new_img_width, new_img_height))

        # PASTE FOOTER ON BOTTOM
        y_position = new_img_height - footer_height
        img_resized.paste(footer, (0, y_position), footer)

        output_path = os.path.join(output_folder, filename)
        img_resized.convert("RGB").save(output_path, quality=100)

        print("Saved:", output_path)

print("DONE — All photos processed!")
```
